```python
import pandas as pd
salary1=pd.read_csv("Salary_Data.csv")
salary1
```

```python
from matplotlib import pyplot as plt
import pandas as pd

x=salary1['YearsExperience']
y=salary1['Salary']

plt.plot(x,y)
plt.show()
```

```python
import seaborn as sns
sns.catplot(x="YearsExperience", y="Salary",data=salary1)
```

```python
y = salary1[['Salary']]             #dependant variable
x = salary1[['YearsExperience']]    #independent variable

from sklearn.linear_model import LinearRegression               #for linear regression
from sklearn.model_selection import train_test_split as tts     #for splitting train and test 

x_train,x_test,y_train,y_test = tts(x,y,test_size=0.3)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)

y_test.head()
```

```python
y_pred[0:5]
```

```python
from sklearn.metrics import mean_squared_error as mse

mse(y_test,y_pred)**0.5           # x**0.5 stands for x raised to the power 0.5
```

```python
from matplotlib import pyplot as plt

plt.scatter(x_test,y_test,color='r',linewidth=2)
plt.plot(x_test,y_pred,color='g',linewidth=2)

plt.title("Red-Test Values | Green-Predicted Values")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

```python
model_score = lr.score(x_test,y_test)
print(model_score*100,'%')
```
